# Logistic Regression

- predict a categorical dependent variable from umber of independent variable
- Logistic regression curve look like a sigmoid curve





## `Maximum Likelihood`
- Product of probabliies of happening of all categories:
    - for example if there are 4 samples and 2 are in 'YES' cat and 2 are in 'NO' cat then maximum likelihood will be 0.53*0.84*(1-0.03)*(1-0.43)

## `Best Fit Curve`
- The best fit curve will be decided after getting the maximum likelihood for the sigmoid curve

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
dataset = pd.read_csv(r"your_datafile_path")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
X.shape
```

```python
print(X,y)
```

```python
#split data for train,test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=0.25, random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train,X_test)
```

```python
#logistic model prepn
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
print(X_train,y_train)
```

```python
#predicting new feature
print(classifier.predict(sc.transform([[54,19000]])))
```

```python
#predicting test reults
y_pred = classifier.predict(sc.transform(X_test))
```

```python
y_test
```

```python
comparison=pd.DataFrame(y_test,classifier.predict(sc.transform(X_test)))
```

```python
print(comparison)
```

```python
#ConfusionMatrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
as1 = accuracy_score(y_test,y_pred)
print(cm,as1)
```
